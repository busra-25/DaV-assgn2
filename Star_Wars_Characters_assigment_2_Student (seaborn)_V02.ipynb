Seaborn is a Python data visualization library based on matplotlib (it is the go to library for plotting in Python). It provides a high-level interface for drawing **fascinating, attractive and informative graphics**. Moreover, Seaborn is a library for making **statistical graphics** in Python. It builds on top of matplotlib and integrates closely with pandas data structures.

Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. Its dataset-oriented, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them

**This can now be plotted much more concisely.**<br>
Axes.bar_label() automatically labels bars [since matplotlib 3.4](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar_label.html)

```python
fig, ax = plt.subplots()
colors= plt.get_cmap('tab10').colors
axes = ax.bar(df['gender'].value_counts().index, df['gender'].value_counts().values, color =colors[:len(df['gender'].value_counts())] )
ax.bar_label(axes )
ax.set_title('Gender', color='blue');
```

```python
#with seaborn barplot
gender_counts = df['gender'].value_counts().sort_values(ascending=False)
fig, ax = plt.subplots()
sns.barplot(x = gender_counts.index, y=gender_counts.values, palette='tab10')
for i in ax.containers:
    ax.bar_label(i)

plt.xticks(rotation=45);
```

```python
# with seaborn countplot
fig, ax = plt.subplots()
sns.countplot(x='gender', data= df,order = df.gender.value_counts().index, palette='tab10');
for i in ax.containers:
    ax.bar_label(i)
```

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">TASK_3 - Plotting the MASS under 3 groups: "High_mass", "Normal_mass", "Low_mass" </p>


<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
df.sample()
```

```python
df['mass'].isna().value_counts()
```

```python
df0 = df.copy()
```

```python
df0['mass'] = pd.to_numeric(df0['mass'], errors='coerce')
```

```python
df0["mass_cat"] = df0.mass.apply(lambda x : "High_mass" if x >=100.0 else  ("Normal_mass" if x> 50.0 and x<100.0 else "Low_mass" if x<=50  else np.nan ))
```

```python
#mass_cat
df0['mass_cat'].value_counts()
```

```python
# with seaborn barplot
fig, ax = plt.subplots(figsize=(8,6))

sns.barplot(x= df0['mass_cat'].value_counts().index, y=df0['mass_cat'].value_counts().values,  palette='tab10' )
ax.set_title('Mass Groups' , color='blue')
for i in ax.containers:
    ax.bar_label(i, color='red');
```

**Seaborn color palettes**

sns.color_palette("bone", n_colors=20)


https://seaborn.pydata.org/tutorial/color_palettes.html
https://www.practicalpythonfordatascience.com/ap_seaborn_palette

```python
# with seaborn countplot
fig, ax = plt.subplots()
sns.countplot(x='mass_cat', data=df0, fill=True, palette='OrRd' )
ax.set_facecolor('lightblue')
for i in ax.containers:
    ax.bar_label(i, color='red');
```
